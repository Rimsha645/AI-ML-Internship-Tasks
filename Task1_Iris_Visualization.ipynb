```python
# Task 1: Exploring and Visualizing the Iris Dataset (CSV version)

# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load Dataset using pandas
df = sns.load_dataset('iris')

# Step 3: Basic Inspection
print("Dataset Shape:", df.shape)
print("\nColumn Names:", df.columns.tolist())
print("\nFirst 5 Rows:\n", df.head())

# Step 4: Summary Info
print("\nDataset Info:")
print(df.info())

# Step 5: Statistical Summary
print("\nSummary Statistics:")
print(df.describe())

# Step 6: Scatter Plot - Feature Relationships
sns.pairplot(df, hue='species')
plt.suptitle("Scatter Plot (Pairplot) of Iris Features by Species", y=1.02)
plt.show()

# Step 7: Histograms - Value Distributions
df.hist(figsize=(10, 8), edgecolor='black')
plt.suptitle("Histograms of Iris Features", fontsize=14)
plt.tight_layout()
plt.show()

# Step 8: Box Plots - Outlier Detection
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title("Boxplot of Iris Features")
plt.show()
```
